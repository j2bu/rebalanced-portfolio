# Returns & Classification

### Step 1. 每个subfund & ETF 每一期的return
### Step 2. 每个subfund & ETF 滚动corr
### Step 3. 每个fund每一期前三corr & excess return > 0 subfund => good subfund

```python
subfunds_sector = form13f_FF4.subfund_sector.unique()
subfunds_sector
```

```python
subfunds_vs = form13f_FF4.subfund_vs.unique()
subfunds_vs
```

```python
iPERIOD_ENDs = form13f_FF4.iPERIOD_END.unique()
iPERIOD_ENDs
```

```python
iCIKs = form13f_FF4.iCIK.unique()
iCIKs
```

### By Sector

```python
groups = form13f_FF4.groupby(["subfund_sector", 'iPERIOD_END'])
```

```python
groups.get_group(('7195_Materials', datetime.date(2013, 6, 30)))
```

```python
returns_sector = pd.DataFrame(columns = ["iCIK", "subfund_sector", "iPERIOD_END", "return", "sector_return"])
for i in subfunds_sector:
    iCIK = i.split("_")[0]
    for j in iPERIOD_ENDs:
        try:
            group = groups.get_group((i,j))
            sums = group.sum()
            r = (sums['iMARKET_VALUE'] - sums['market_V_3mos']) / sums['market_V_3mos']
            r_sec = (sums['sector_close'] - sums['sector_close_3mos']) / sums['sector_close_3mos']
            returns_sector = returns_sector.append({'iCIK': iCIK, 'subfund_sector': i, 'iPERIOD_END':j, 'return':r, 'sector_return': r_sec},
                                     ignore_index = True)
        except KeyError:
            returns_sector = returns_sector.append({'iCIK': iCIK, 'subfund_sector': i, 'iPERIOD_END':j, 'return':'NA', 'sector_return': 'NA'},
                                     ignore_index = True)
            
returns_sector
```

```python
returns_sector = returns_sector[returns_sector["return"] != 'NA']
returns_sector
```

```python
g = returns_sector.groupby("subfund_sector")
```

```python
corr_sector = pd.DataFrame(columns = ["iCIK", "subfund_sector", "iPERIOD_END", "corr_sector"])

for i in subfunds_sector:
    group = g.get_group(i)
    df = pd.DataFrame(columns = ["iCIK", "subfund_sector", "iPERIOD_END", "corr_sector"])
    df["iCIK"] = group["iCIK"]
    df["subfund_sector"] = group["subfund_sector"]
    df["iPERIOD_END"] = group["iPERIOD_END"]
    
    corrs = []
    for k in range(len(group)):
        history_corr = group.iloc[:k,3:].astype(float).corr()["return"]["sector_return"]
        corrs.append(history_corr)
        
    df["corr_sector"] = corrs
    corr_sector = corr_sector.append(df)
corr_sector
```

```python
corr_sector = pd.merge(returns_sector, corr_sector, on=["iCIK",'subfund_sector','iPERIOD_END'])
corr_sector["good_subfund_sector"] = 0
corr_sector
```

```python
corr_sector = corr_sector.sort_values(["iCIK", "iPERIOD_END"])
corr_sector
```

```python
df = pd.DataFrame(columns = ["iCIK", "subfund_sector", "iPERIOD_END", "corr_sector", 'good_subfund_sector'])
cg = corr_sector.groupby(["iCIK", "iPERIOD_END"])
for i in iCIKs:
    for p in iPERIOD_ENDs:
        try:
            g = cg.get_group((str(i),p))
            for index in g['corr_sector'].nlargest(3).index:
                if(g.loc[index, 'return'] > g.loc[index, 'sector_return']):
                    g.loc[index, 'good_subfund_sector'] = 1
            
            df = df.append(g)
        except KeyError:
            continue
    
df
```

```python
df["iCIK"] = df["iCIK"].astype(int)
form13f_good_subfund_sec = pd.merge(form13f_FF4, df, on=["iCIK", "subfund_sector", "iPERIOD_END"])
form13f_good_subfund_sec
```
