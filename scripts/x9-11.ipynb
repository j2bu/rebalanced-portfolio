```python
########################################## testing ##########################################
stock_data = good_subfund_labels[good_subfund_labels.iCUSIP == '219350105']
stock_data
```

```python
stock_data_good = stock_data[stock_data.good_subfund_vs == 1]
stock_data_good
```

```python
good_subfund_labels[(good_subfund_labels.iPERIOD_END == datetime.date(2014,9,30))
                       & (good_subfund_labels.subfund_vs == '7195_Large-Cap Value')]
```

```python
weights = []
for index, row in stock_data_good.iterrows():
    stock_market_value = row.iMARKET_VALUE
    good_subfund_data = good_subfund_labels[(good_subfund_labels.iPERIOD_END == row.iPERIOD_END)
                       & (good_subfund_labels.subfund_vs == row.subfund_vs)]
    weight = stock_market_value / good_subfund_data.sum().iMARKET_VALUE
    weights.append(weight)
np.array(weights).mean()
```

```python
###########################################################################################
```

```python
iPERIOD_END = []
iCUSIP = []
mean_weights = []
for period in iPERIOD_ENDs:
    for stock in all_stocks:
        period_data = good_subfund_labels[good_subfund_labels.iPERIOD_END == period]
        stock_data = period_data[period_data.iCUSIP == stock]
        stock_data_good = stock_data[stock_data.good_subfund_vs == 1]
    
        weights = []
        for index, row in stock_data_good.iterrows():
            stock_market_value = row.iMARKET_VALUE
            good_subfund_data = good_subfund_labels[(good_subfund_labels.iPERIOD_END == row.iPERIOD_END)
                           & (good_subfund_labels.subfund_vs == row.subfund_vs)]
            weight = stock_market_value / good_subfund_data.sum().iMARKET_VALUE
            weights.append(weight)
        iPERIOD_END.append(period)
        iCUSIP.append(stock)
        mean_weights.append(np.array(weights).mean())

x11_df = pd.DataFrame({"iCUSIP": iCUSIP, "iPERIOD_END": iPERIOD_END, "x11": mean_weights})
x11_df  
```

```python
x11_df[x11_df.iPERIOD_END == datetime.date(2016,6,30)]
```
