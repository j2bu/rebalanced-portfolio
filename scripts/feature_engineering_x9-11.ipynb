```python
weights = []
for index, row in stock_data_good.iterrows():
    stock_market_value = row.iMARKET_VALUE
    good_subfund_data = good_subfund_labels[(good_subfund_labels.iPERIOD_END == row.iPERIOD_END)
                       & (good_subfund_labels.subfund_vs == row.subfund_vs)]
    weight = stock_market_value / good_subfund_data.sum().iMARKET_VALUE
    weights.append(weight)
np.array(weights).mean()
```

```python
###########################################################################################
```

```python
mean_weights = []
for stock in all_stocks:
    stock_data = good_subfund_labels[good_subfund_labels.iCUSIP == stock]
    stock_data_good = stock_data[stock_data.good_subfund_vs == 1]
    
    weights = []
    for index, row in stock_data_good.iterrows():
        stock_market_value = row.iMARKET_VALUE
        good_subfund_data = good_subfund_labels[(good_subfund_labels.iPERIOD_END == row.iPERIOD_END)
                       & (good_subfund_labels.subfund_vs == row.subfund_vs)]
        weight = stock_market_value / good_subfund_data.sum().iMARKET_VALUE
        weights.append(weight)
    mean_weights.append(np.array(weights).mean())

x11_df = pd.DataFrame({"iCUSIP": all_stocks, "x11": mean_weights})
x11_df  
```

```python
x9_10_df = pd.merge(x9_df, x10_df, on = "iCUSIP")
x9_10_11_df = pd.merge(x9_10_df, x11_df, on = "iCUSIP")
x9_10_11_df
```
