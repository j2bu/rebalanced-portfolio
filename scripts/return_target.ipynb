```python
import pandas as pd
import datetime
from dateutil.relativedelta import *
import numpy as np
```

```python
fund_prices_FF = pd.read_csv("../output/fund_prices_FF.csv", sep = "\t")
fund_prices_FF["pSP_DATE"]= pd.to_datetime(fund_prices_FF["pSP_DATE"]).dt.date
fund_prices_FF
```

```python
good_subfund_labels = pd.read_csv("../output/good_subfund_labels.csv")
good_subfund_labels["iPERIOD_END"] = pd.to_datetime(good_subfund_labels["iPERIOD_END"]).dt.date
good_subfund_labels
```

```python
all_stocks = good_subfund_labels.iCUSIP.unique()
print(len(all_stocks))
all_stocks
```

```python
iPERIOD_ENDs1 = good_subfund_labels.iPERIOD_END.unique()
iPERIOD_ENDs1
```

```python
iPERIOD_ENDs2 = iPERIOD_ENDs1 + relativedelta(days=+45)
iPERIOD_ENDs2
```

```python
for period in iPERIOD_ENDs2:
    if period not in fund_prices_FF.pSP_DATE.unique():
        print(period)
```

```python
for i in range(len(iPERIOD_ENDs2)):
    while iPERIOD_ENDs2[i] not in fund_prices_FF.pSP_DATE.unique():
        iPERIOD_ENDs2[i] = iPERIOD_ENDs2[i] + relativedelta(days=+1)
iPERIOD_ENDs2
```

```python
for period in iPERIOD_ENDs2:
    if period not in fund_prices_FF.pSP_DATE.unique():
        print(period)
```

```python
fund_prices_FF[(fund_prices_FF.pSP_CUSIP == '219350105') & (fund_prices_FF.pSP_DATE == datetime.date(2016,6,30))]["pSP_CLOSE"].values[0]
```

```python
iCUSIPs = []
iPERIOD_ENDs = []
period_nows = []
period_nexts = []
returns = []

counter = 0
for stock in all_stocks:
    counter += 1
    print("stock", counter)
    for i in range(len(iPERIOD_ENDs1)-1):
        try:
            iPERIOD_END = iPERIOD_ENDs1[i]
            period_now = iPERIOD_ENDs2[i]
            period_next = iPERIOD_ENDs2[i+1]
            price_now = fund_prices_FF[(fund_prices_FF.pSP_CUSIP == stock) & (fund_prices_FF.pSP_DATE == period_now)]["pSP_CLOSE"].values[0]
            price_next = fund_prices_FF[(fund_prices_FF.pSP_CUSIP == stock) & (fund_prices_FF.pSP_DATE == period_next)]["pSP_CLOSE"].values[0]
            r = (price_next - price_now) / price_now
        except IndexError:
            r = np.nan
        
        iCUSIPs.append(stock)
        iPERIOD_ENDs.append(iPERIOD_END)
        period_nows.append(period_now)
        period_nexts.append(period_next)
        returns.append(r)

returns_df = pd.DataFrame({"iCUSIP":iCUSIPs, "iPERIOD_END":iPERIOD_ENDs, "period_now":period_nows, 
                           "period_next":period_nexts, "return":returns})
returns_df
```

```python
returns_df.to_csv("../output/return_target.csv", index=False)
```
